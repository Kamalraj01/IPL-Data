```python
# calculate man of the match 
plt.figure(figsize=(6,3))
man_of_the_match=df['player_of_the_match'].value_counts().nlargest(10)
man_of_the_match
man_of_the_match.plot(kind="bar",color='g')
plt.show()
```

```python
# Highest score in venue wise
max_run=df.groupby('venue').first_ings_score.max()
plt.figure(figsize=(6,4))
max_run.plot(kind="bar",color="blue")
plt.ylabel("Runs")

plt.show()
```

```python
# Teamwise points table
points_table=df.match_winner.value_counts()
points_table.plot(kind="bar",color='magenta')
plt.xlabel('Teams')
plt.ylabel('Number of wins')
plt.show()
```

```python
# Who won the toss how many times
toss_winner=df.toss_winner.value_counts()
toss_winner.plot(kind="bar",color="g")
plt.xlabel("Teams")
plt.ylabel("Number of toss won")
plt.show()
```

```python
# Toss Decision
plt.figure(figsize=(6,5))
toss_decision=df.toss_decision.value_counts()
toss_decision.plot(kind="pie",color="g",autopct="%0.2f%%")
plt.legend(loc='upper right')
plt.show()
```

```python
# Top 20 Best Bowling
plt.figure(figsize=(7,7))
bowling=df.best_bowling.value_counts().nlargest(20)
bowling.plot(kind="barh",color='magenta')
plt.show()
```

```python
# Top 10 Best Bowling Figure
plt.figure(figsize=(8,8))
bowling_fig=df.groupby('best_bowling').best_bowling_figure.value_counts().nlargest(10)
bowling_fig.plot(kind="barh",color='g')
plt.ylabel('Best Bowling Figure')
plt.show()
```

```python
# Top scorer in ipl 2022
topscore=df.groupby('top_scorer')['highscore'].max().nlargest(10)

topscore_1=topscore.reset_index()

plt.figure(figsize=(8,6))
sns.barplot(x='top_scorer',y='highscore',data=topscore_1,palette='viridis')

for index,row in topscore_1.iterrows():
    plt.text(index,row['highscore'],str(row['highscore']),ha='center',va='bottom')
plt.title('Top Scorers in IPl 2022')
plt.xlabel('Top Scorers')
plt.ylabel('Highest score')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
# Best Bowlers

topbowlers = df['best_bowling'].value_counts().nlargest(10)

topbowlers = topbowlers.reset_index()
topbowlers.columns =['Bowler','Occurrences']

plt.figure(figsize=(8,6))
sns.barplot(x='Bowler',y='Occurrences',data=topbowlers,palette='viridis')

for index,row in topbowlers.iterrows():
    plt.text(index,row['Occurrences'],str(row['Occurrences']),ha='center',va='bottom')

plt.title('Top Bowlers in IPl 2022')
plt.xlabel('Bowler')
plt.ylabel('Occurrences')
plt.xticks(rotation=45,ha='right')
plt.tight_layout()
plt.show()
```

```python
# First Innings Sscores
plt.figure(figsize=(8,6))
sns.histplot(df['first_ings_score'],bins=10,kde=True,color='g')
plt.title('First Innings Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# Matched played in Each venue
plt.figure(figsize=(8,6))
venue_counts = df['venue'].value_counts()
ax=venue_counts.plot(kind='bar',color='g')

plt.title('Matches Played in Each Venue')
plt.xlabel('Venue')
plt.ylabel('Number of Matches')
plt.show()
```
